1. Видатні персоналії
- ***Жан Батіст Уедраого*** - колишній президент Верхньої Вольти (з 1984 Буркіна-Фасо). Є доктором медицини, нині працює у клініці на околиці міста Уагадугу.
- ***Норберт Зонго*** - журналіст і письменник із західноафриканської держави Буркіна-Фасо. Норберт Зонго відвідував журналістські школи у Того та Камеруні. Він працював в урядовій газеті «Sidwaya» та приватних газетах.
- ***Сей Зербо*** - державний та військовий діяч Верхньої Вольти (нині Буркіна-Фасо), президент і прем'єр-міністр у 1980–1982 роках, полковник. Зербо отримував державну пенсію як колишній глава держави та полковник армії у відставці.
2. Пам'ятки культури та природи
- **Національний Музей**. Тут демонструються традиційні маски, шматки глиняного посуду, інші ритуальні речі головних етнічних груп країни.
- **Собор**, розташований біля Палацу Моро-Наба, є найбільшою церквою в Західній Африці.
- **Великий Ринок Уагадугу**, що знаходиться в центрі міста і має безліч великих декоративних башточок по верху фасаду, також варто відвідування.
- **Озеро Teгреакуте**, приємне місце, щоб зупинитися і відпочити на лоні природи.

```python
def igcd(num1, num2):
    num1, num2 = abs(num1), abs(num2)
    while num1 != num2:
        if num1 > num2:
            num1 -= num2
        else:
            num2 -= num1
            
    return num1


def task_2():
    lines = input().split(" ")
    nums = []
    for i in lines:
        try:
            nums.append(int(i))
        except ValueError:
            return f"Error! '{i}' isn't integer number!"
        
    if len(nums) != 2:
        return "Error! There must be two numbers!"
        
    num1, num2 = nums
    
    return f"The greatest common divisor of {num1} and {num2} is {igcd(num1, num2)}"
```
